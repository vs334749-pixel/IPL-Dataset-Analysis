```python
print("Missing values before cleaning:")
print(df.isnull().sum())
```

```python
df["Runs"] = df["Runs"].fillna(df["Runs"].mean())
df["Wickets"] = df["Wickets"].fillna(df["Wickets"].mean())
df["Overs"] = df["Overs"].fillna(df["Overs"].mean())
df["Strike_Rate"] = df["Strike_Rate"].fillna(df["Strike_Rate"].mean())
```

# **The 10 Questions and Answers**

# 1. Which team has the highest average strike rate and what is its average runs?

```python
import pandas as pd

df = pd.read_csv("IPL_Dataset_Auto_Cleaned.csv")

result = df.groupby("Team").agg({
    "Strike_Rate":"mean",
    "Runs":"mean"
}).sort_values("Strike_Rate",ascending=False)

print(result.head(1))
```

# 2. Find the Top 5 players with the highest average strike rate.

```python
top_players = df.groupby("Player_of_Match")["Strike_Rate"] \
.mean() \
.sort_values(ascending=False) \
.head(5)

print(top_players)
```

# 3. Which team wins most matches at each venue?

```python
result = df.groupby(["Venue","Winner"]).size()

print(result)
```

# 5. Find matches where Runs are above average AND Strike Rate is above average.

```python
result = df[
(df["Runs"]>df["Runs"].mean()) &
(df["Strike_Rate"]>df["Strike_Rate"].mean())
]

print(result)
```

# 6.Find the Top 3 Teams with the Highest Average Strike Rate and More Than 2 Matches

```python
import pandas as pd

df = pd.read_csv("IPL_Dataset_Auto_Cleaned.csv")

result = (
    df.groupby("Team")
      .agg(
          Matches=("Match_ID", "count"),
          Avg_Strike_Rate=("Strike_Rate", "mean")
      )
)

result = result[result["Matches"] > 2]

print(result.sort_values("Avg_Strike_Rate", ascending=False).head(3))
```

# 7.Find the Team That Has Won the Most Matches at Each Venue

```python
import pandas as pd

df = pd.read_csv("IPL_Dataset_Auto_Cleaned.csv")

result = (
    df.groupby(["Venue", "Winner"])
      .size()
      .reset_index(name="Wins")
)

idx = result.groupby("Venue")["Wins"].idxmax()

print(result.loc[idx])
```

# 8.Find Teams Whose Average Runs Are Greater Than the Overall Average Runs

```python
import pandas as pd

df = pd.read_csv("IPL_Dataset_Auto_Cleaned.csv")

overall_avg = df["Runs"].mean()

team_avg = df.groupby("Team")["Runs"].mean()

print(team_avg[team_avg > overall_avg])
```

# 9.Create a New Performance Index and Find the Best Match

```python
import pandas as pd

df = pd.read_csv("IPL_Dataset_Auto_Cleaned.csv")

df["Performance_Index"] = (
    df["Runs"] * df["Strike_Rate"]
) / df["Overs"]

best_match = df.loc[df["Performance_Index"].idxmax()]

print(best_match)
```

# 10.Generate a Complete Team Performance Dashboard-

```python
import pandas as pd

df = pd.read_csv("IPL_Dataset_Auto_Cleaned.csv")

dashboard = df.groupby("Team").agg({
    "Runs": ["sum", "mean", "max", "min"],
    "Wickets": ["sum", "mean"],
    "Overs": ["mean"],
    "Strike_Rate": ["mean", "max"],
    "Match_ID": "count"
})

print(dashboard)
```
