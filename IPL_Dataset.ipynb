Fill Missing Values Automatically

# Fill numerical columns with mean

```python
print("Missing values before cleaning:")
print(df.isnull().sum())
```

```python
df["Runs"] = df["Runs"].fillna(df["Runs"].mean())
df["Wickets"] = df["Wickets"].fillna(df["Wickets"].mean())
df["Overs"] = df["Overs"].fillna(df["Overs"].mean())
df["Strike_Rate"] = df["Strike_Rate"].fillna(df["Strike_Rate"].mean())
```

```python
print("\nMissing values after cleaning:")
print(df.isnull().sum())
```

```python
df.to_csv("IPL_Dataset_Auto_Cleaned.csv", index=False)
```
